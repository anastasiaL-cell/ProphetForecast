```python
import joblib

model = joblib.load("tuned_xgboost.pkl")
model
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("feature_engineered.csv")

df["date"] = pd.to_datetime(df["date"])
df = df.set_index("date")
df = df.dropna()

df.head()
```

```python
df.columns
```

```python
last_row = df.drop(columns=["unit_sales"]).iloc[-1:]

last_row
```

```python
prediction = model.predict(last_row)

print(prediction)
```

```python
def predict_latest_sales():
    last_row = df.drop(columns=["unit_sales"]).iloc[-1:]
    prediction = model.predict(last_row)[0]
    return prediction

latest_prediction = predict_latest_sales()

print(f"Predicted unit sales: {latest_prediction:.2f}")
```
